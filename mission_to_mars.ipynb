```python
# Path to chromedriver
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
# Fire it up
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Read desired URL
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Search just like we did with `results`
titles = soup.find_all('div', class_="content_title")
paragraphs = soup.find_all('div', class_="article_teaser_body")

# We need to offset the title and paragraphs by 1 because the first title isn't for an article and has no paragraph
for x in range(len(titles)-1):
    print('-'*70)
    print('TITLE: ', titles[x+1].text)
    try:
        print('TEASER: ', paragraphs[x].text)
    except:
        pass
```

# Mars Hemispheres

```python
# Read desired URL
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')
```

```python
soup.find_all('div', class_='item')
```

```python
# Retreive all 4 results
results = soup.find_all('div', class_='item')

# Loop through the items previously stored
hemisphere_image_urls = []
base_url = 'https://astrogeology.usgs.gov'
for result in results: 
    
    # Store title
    title = result.find('h3').text
    
    # Store link that leads to full image website
    image_page_url = result.find('a', class_='itemLink')['href']
    
    # Visit the page that contains the hi-res image
    browser.visit(base_url + image_page_url)
    html = browser.html
    soup = bs(html, 'html.parser')
    
    # Retrieve hi-res image URL
    image_url = base_url + soup.find('img', class_='wide-image')['src']
    
    # Append the retreived information into a list of dictionaries 
    hemisphere_image_urls.append({"title" : title, "img_url" : img_url})
    
    
hemisphere_image_urls
```
